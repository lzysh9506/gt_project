```python
def trade_df_filter(df_name, filter_date):
  df_name.columns = ['TIME_IN', 'TICKET', 'SYMBOL', 'TYPE', 'VOLUME', 'PRICE_IN', 'SL', 'TP', 'TIME_OUT', 'PRICE_OUT', 'COMMISION', 'SWAP', 'PROFIT']
  df_name['TIME_IN'] = pd.to_datetime(df_name['TIME_IN'])
  df_name['TIME_OUT'] = pd.to_datetime(df_name['TIME_OUT'])
  specific_date = pd.to_datetime(filter_date)
  filtered_df = df_name[df_name['TIME_IN'] > specific_date]
  filtered_df['TYPE'] = filtered_df['TYPE'].replace({'buy': 0, 'sell': 1})
  filtered_df = filtered_df.dropna(axis=1, how='all')
  return filtered_df

def compensation_df_filter(df_name, filter_date):
  df_name.columns = ['TIME_IN', 'DEAL', 'SYMBOL', 'TYPE', 'DIRECTION', 'VOLUME', 'PRICE', 'TICKET', 'COMMISION', 'FEE', 'SWAP', 'PROFIT', 'BALANCE', 'COMMENT']
  df_name['TIME_IN'] = pd.to_datetime(df_name['TIME_IN'])
  specific_date = pd.to_datetime(filter_date)
  filtered_df = df_name[df_name['TIME_IN'] > specific_date]
  filtered_df['TYPE'] = filtered_df['TYPE'].replace({'buy': 0, 'sell': 1, 'balance': 2})
  filtered_df = filtered_df.dropna(axis=1, how='all')
  return filtered_df
```

```python
def create_trade_table(table_name, column_def):
  # Create the table query
  create_table_query = f'CREATE TABLE {table_name} ({column_def})'

  # Execute the create table query
  cursor.execute(create_table_query)

def insert_trade_values(table_name, df_name, column_name):
  data_to_insert = df_name[column_name].values.tolist()
  column_str = ', '.join(column_name)
  insert_query = f"INSERT INTO {table_name} ({column_str}) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"

  # Execute the INSERT query
  cursor.executemany(insert_query, data_to_insert)

  # Commit the changes
  cnx.commit()

def create_compensation_table(table_name, column_def):
  # Create the table query
  create_table_query = f'CREATE TABLE {table_name} ({column_def})'

  # Execute the create table query
  cursor.execute(create_table_query)

def insert_compensation_values(table_name, df_name, column_name):
  data_to_insert = df_name[column_name].values.tolist()
  column_str = ', '.join(column_name)
  insert_query = f"INSERT INTO {table_name} ({column_str}) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"

  # Execute the INSERT query
  cursor.executemany(insert_query, data_to_insert)

  # Commit the changes
  cnx.commit()
```

```python
query = '''
select
  a.TIME_IN as TIME_IN_A, a.TICKET as TICKET_A, a.TYPE as TYPE_A, a.VOLUME as VOLUME_A, a.PRICE_IN as PRICE_IN_A, a.PRICE_OUT as PRICE_OUT_A, a.COMMISION as COMMISION_A, a.SWAP as SWAP_A, (a.PRICE_OUT - a.PRICE_IN)*100 as SPREAD_A, a.PROFIT as PROFIT_A,
  b.TIME_IN as TIME_IN_B, b.TICKET as TICKET_B, b.TYPE as TYPE_B, b.VOLUME as VOLUME_B, b.PRICE_IN as PRICE_IN_B, b.PRICE_OUT as PRICE_OUT_B, b.SWAP as SWAP_B, (b.PRICE_OUT - b.PRICE_IN)*100 as SPREAD_B, b.PROFIT as PROFIT_B,
  COALESCE(comp1.COMPENSATION, 0) as COMPENSATION,
  case when b.PROFIT is null then a.PROFIT + a.COMMISION + a.SWAP
  when a.PROFIT < 0 and b.PROFIT > 0 then (a.PROFIT + a.COMMISION + a.SWAP) + (b.PROFIT + b.SWAP) + COALESCE(comp1.COMPENSATION, 0)
  else a.PROFIT + a.COMMISION + a.SWAP + (b.PROFIT + b.SWAP) end as FINAL_PROFIT
from
(
  select *, UNIX_TIMESTAMP(TIME_IN) as ts
  from {table_name_a}
)as a
left join
(
  select *, UNIX_TIMESTAMP(TIME_IN) as ts
  from {table_name_b}
)as b
on abs(a.TICKET - b.TICKET) <= 30 or abs(a.ts - b.ts) <= 2*60
left join
(
  select UNIX_TIMESTAMP(TIME_IN) as ts, PROFIT as COMPENSATION, CAST(REGEXP_SUBSTR(COMMENT, '[0-9]+') AS UNSIGNED) as COMPENSATE_TRADE_ID
  from {table_comp_name}
  where COMMENT like '%compensation%'
) as comp1
on comp1.COMPENSATE_TRADE_ID = a.TICKET
union
select
  c.TIME_IN as TIME_IN_A, c.TICKET as TICKET_A, c.TYPE as TYPE_A, c.VOLUME as VOLUME_A, c.PRICE_IN as PRICE_IN_A, c.PRICE_OUT as PRICE_OUT_A, c.COMMISION as COMMISION_A, c.SWAP as SWAP_A, (c.PRICE_OUT - c.PRICE_IN)*100 as SPREAD_A, c.PROFIT as PROFIT_A,
  d.TIME_IN as TIME_IN_B, d.TICKET as TICKET_B, d.TYPE as TYPE_B, d.VOLUME as VOLUME_B, d.PRICE_IN as PRICE_IN_B, d.PRICE_OUT as PRICE_OUT_B, d.SWAP as SWAP_B, (d.PRICE_OUT - d.PRICE_IN)*100 as SPREAD_B, d.PROFIT as PROFIT_B,
  COALESCE(comp2.COMPENSATION, 0) as COMPENSATION,
  case when c.PROFIT is null then d.PROFIT + d.COMMISION + d.SWAP
  when c.PROFIT < 0 and d.PROFIT > 0 then (c.PROFIT + c.COMMISION + c.SWAP) + (d.PROFIT + d.SWAP) + COALESCE(comp2.COMPENSATION, 0)
  else c.PROFIT + c.COMMISION + c.SWAP + (d.PROFIT + d.SWAP) end as FINAL_PROFIT
from
(
  select *, UNIX_TIMESTAMP(TIME_IN) as ts
  from {table_name_a}
)as c
right join
(
  select *, UNIX_TIMESTAMP(TIME_IN) as ts
  from {table_name_b}
)as d
on abs(c.TICKET - d.TICKET) <= 30 or abs(c.ts - d.ts) <= 2*60
left join
(
  select UNIX_TIMESTAMP(TIME_IN) as ts, PROFIT as COMPENSATION, CAST(REGEXP_SUBSTR(COMMENT, '[0-9]+') AS UNSIGNED) as COMPENSATE_TRADE_ID
  from {table_comp_name}
  where COMMENT like '%compensation%'
) as comp2
on comp2.COMPENSATE_TRADE_ID = c.TICKET

'''.format(table_name_a = table_name_a, table_name_b = table_name_b, table_comp_name = table_comp_name)
cursor.execute(query)
```
